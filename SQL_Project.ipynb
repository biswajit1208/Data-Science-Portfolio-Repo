### Problem 1: Find the total number of crimes recorded in the CRIME table.

```python
pd.read_sql('select count(*) FROM chicago_school.chicagocrimedata',mydb)
```

### Problem 2: List community areas with per capita income less than 11000.

```python
pd.read_sql('select COMMUNITY_AREA_NAME, COMMUNITY_AREA_NUMBER \
            FROM chicago_school.chicagocensusdata \
            WHERE PER_CAPITA_INCOME < 11000',mydb)
```

### Problem 3: List all case numbers for crimes involving minors?

```python
pd.read_sql("""SELECT CASE_NUMBER 
            FROM chicago_school.chicagocrimedata 
            WHERE DESCRIPTION LIKE '%minor%' """,mydb)
```

### Problem 4: List all kidnapping crimes involving a child?(children are not considered minors for the purposes of crime analysis)

```python
pd.read_sql("""SELECT * 
            FROM chicago_school.chicagocrimedata 
            WHERE PRIMARY_TYPE = 'kidnapping' """,mydb)
```

### Problem 5: What kind of crimes were recorded at schools?

```python
pd.read_sql("""SELECT DISTINCT(PRIMARY_TYPE)
            FROM chicago_school.chicagocrimedata
            WHERE LOCATION_DESCRIPTION LIKE '%SCHOOL%'""",mydb)
```

### Problem 6: List the average safety score for all types of schools.

```python
pd.read_sql("""SELECT "Elementary, Middle, or High School ,AVG(SAFETY_SCORE) AS Avg_SAFETY_SCORE 
FROM chicago_school.chicagopublicschools
GROUP BY Elementary, Middle, or High School """,mydb)
```

### Problem 7: List 5 community areas with highest % of households below poverty line.

```python
pd.read_sql("""SELECT COMMUNITY_AREA_NAME,PERCENT_HOUSEHOLDS_BELOW_POVERTY
FROM chicago_school.chicagocensusdata
ORDER BY PERCENT_HOUSEHOLDS_BELOW_POVERTY DESC
LIMIT 5""",mydb)
```

### Problem 8: Which community area(number) is most crime prone?

```python
pd.read_sql("""SELECT cr.COMMUNITY_AREA_NUMBER,COUNT(cr.COMMUNITY_AREA_NUMBER) AS FREQUENCY
FROM chicago_school.chicagocrimedata AS cr
GROUP BY cr.COMMUNITY_AREA_NUMBER
ORDER BY COUNT(COMMUNITY_AREA_NUMBER) DESC
LIMIT 1""",mydb)
```

### Problem 9: Use a sub-query to find the name of the community area with highest hardship index.

```python
pd.read_sql("""SELECT COMMUNITY_AREA_NAME
FROM chicago_school.chicagocensusdata
WHERE HARDSHIP_INDEX = (SELECT MAX(HARDSHIP_INDEX) FROM chicago_school.chicagocensusdata)""",mydb)
```

### Problem 10: Use a sub-query to determine the Community Area Name with most number of crimes?

```python
pd.read_sql("""SELECT COMMUNITY_AREA_NAME 
FROM chicago_school.chicagocensusdata
WHERE COMMUNITY_AREA_NUMBER = (
SELECT cr.COMMUNITY_AREA_NUMBER
FROM chicago_school.chicagocrimedata AS cr
GROUP BY cr.COMMUNITY_AREA_NUMBER
ORDER BY COUNT(cr.COMMUNITY_AREA_NUMBER) DESC
LIMIT 1 )""",mydb)
```

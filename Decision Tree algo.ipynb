## 1. Import libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
%matplotlib inline
```

```python
dat= pd.read_csv('B:\Interview prep\data\iris_dataset.csv')
```

## 5. Handling Catagorical variables

```python
from sklearn.preprocessing import LabelEncoder
labelencoder=LabelEncoder()
dat['Species']=labelencoder.fit_transform(dat['Species'])
```

```python
dat
```

## 6. Spliting dataset into expanatory variables and target variable

```python
x=dat[['Sepal.Length','Sepal.Width','Petal.Length','Petal.Width']]
y=dat[['Species']]
```

```python
x
```

```python
y
```

## 7. Spliting explanatory variables and target variable into train set and test set

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.25, random_state = 1)
```

```python
x_train
```

```python
x_test
```

## 8. Decision Tree Classifier with criterion entropy

```python
from sklearn.tree import DecisionTreeClassifier
classifier = DecisionTreeClassifier(criterion = 'entropy', random_state = 0)
classifier.fit(x_train, y_train)
```

## 9. Predict Value

```python
y_pred = classifier.predict(x_test)
y_pred
```

## 10. Confusion Matrix

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print("Confusion Matrix\n\n",cm)
```

## 11. Check for overfitting and underfitting

```python
# print the scores on training and test set

print('Training set score: {:.4f}'.format(classifier.score(x_train, y_train)))

print('Test set score: {:.4f}'.format(classifier.score(x_test, y_test)))
```

## 12.Visualisation of Iris_Dataset by Decision Tree

```python
from sklearn import tree
fn=['sepal length (cm)','sepal width (cm)','petal length (cm)','petal width (cm)']
cn=['setosa', 'versicolor', 'virginica']
fig, axes = plt.subplots(nrows = 1,ncols = 1,figsize = (12,10), dpi=600)
tree.plot_tree(classifier,
               feature_names = fn, 
               class_names=cn,
               filled=True)
fig.savefig('tree.png')
```
